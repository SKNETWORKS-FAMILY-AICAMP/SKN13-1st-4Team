```python
import os
import time
import requests
from PIL import Image
from io import BytesIO
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# 🔹 로고 저장할 폴더
save_dir = "images"
os.makedirs(save_dir, exist_ok=True)

# 🔹 수집 대상 보험사 목록
companies = ["삼성화재", "현대해상", "DB손해보험", "메리츠화재", "KB손해보험"]

# 🔹 크롬 옵션 설정 (헤드리스 모드)
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# 🔹 각 보험사별 수집
for company in companies:
    try:
        # 드라이버 매번 새로 시작 (안정성 ↑)
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service, options=options)

        # 구글 이미지 검색 접속
        query = f"{company} 로고"
        url = f"https://www.google.com/search?q={query}&tbm=isch"
        driver.get(url)

        # 이미지 로딩 대기
        wait = WebDriverWait(driver, 10)
        img = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "img.Q4LuWd")))
        img_url = img.get_attribute("src")

        # 이미지 다운로드 및 저장
        response = requests.get(img_url)
        img_file = Image.open(BytesIO(response.content)).convert("RGB")
        filename = f"{save_dir}/{company}.png"
        img_file.save(filename)
        print(f"✅ 저장 완료: {filename}")

    except Exception as e:
        print(f"❌ {company} 이미지 실패: {e}")

    finally:
        driver.quit()
```

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import urllib3

# 인증서 경고 끄기
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# 저장할 디렉토리
save_dir = "insurance_logos"
os.makedirs(save_dir, exist_ok=True)

# 대상 URL
url = "https://carinfo.knia.or.kr/lmxsrv/cnswc/icnyCnswcList.do"
headers = {"User-Agent": "Mozilla/5.0"}

# 요청
response = requests.get(url, headers=headers, verify=False)
soup = BeautifulSoup(response.text, "html.parser")

# 이미지 추출
img_tags = soup.find_all("img")
for idx, img in enumerate(img_tags):
    src = img.get("src")
    img_url = urljoin(url, src)

    # 파일 이름 만들기 (alt 또는 인덱스 기반)
    name = img.get("alt") or f"logo_{idx+1}"
    name = name.replace(" ", "_").replace("/", "_")  # 안전한 파일명
    filename = os.path.join(save_dir, f"{name}.png")

    try:
        img_data = requests.get(img_url, verify=False).content
        with open(filename, "wb") as f:
            f.write(img_data)
        print(f"✅ 저장 완료: {filename}")
    except Exception as e:
        print(f"❌ 저장 실패: {img_url} / {e}")
```
